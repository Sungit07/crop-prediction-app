```python
import pandas as pd
```

```python
df = pd.read_csv('Ifood_new.csv')
df.sample(5)
```

```python
df['ingredients_clean'] = (
    df['ingredients']
    .str.lower()
    .str.replace('[^a-z,]','',regex=True)
)
df.head()   
```

```python
df['ingredient_list'] = df['ingredients_clean'].str.split(',')
df['ingredient_list'] = df['ingredient_list'].apply(lambda x:[i.strip() for i in x])
df.head()
```

```python
# def find_dish(crop):
#     crop = crop.lower()
#     matches = df[df['ingredient_list'].apply(lambda x: any(crop in ing for ing in x))]
#     return matches[['name', 'ingredients_clean', 'img_url', 'state', 'region']]

# def find_dish(crop):
#     crop=crop.lower()
#     match=[]
#     for l,r in df.iterrows():
#         li=r['ingredient_list']
#         if crop in li:
#             match.append(r)
#     return pd.DataFrame(match)
def find_dish(crop):
    crop = crop.lower()
    match = []

    for idx, row in df.iterrows():   # idx=index, row=row data
        ingredients = row['ingredient_list']  # list of ingredients

        # loop through each ingredient in the list
        for ing in ingredients:
            if crop in ing:     # substring check: 'rice' in 'rice flour'
                match.append(row)
                break           # stop checking further ingredients for this row

    return pd.DataFrame(match)
df3 = find_dish("rice")
df3.head()
```

```python
nutrition = [
    'FOOD-DATA-GROUP1.csv',
    'FOOD-DATA-GROUP2.csv',
    'FOOD-DATA-GROUP3.csv',
    'FOOD-DATA-GROUP4.csv',
    'FOOD-DATA-GROUP5.csv'
]
nutr = [pd.read_csv(file) for file in nutrition]
df2 = pd.concat(nutr,ignore_index=True)
df2.head()
```

```python
df2 = df2.drop(columns=[col for col in df2.columns if "Unnamed" in col],errors='ignore')
df2['food'] = df2['food'].str.lower().str.strip()
df2.head()
```

```python
def get_nutr(ingredient):
    ingredient = ingredient.lower().strip()
    match = df2[df2['food'].str.contains(ingredient,na=False)]
    if match.empty:
        return None
    return match.iloc[0]
```

```python
for dish in df3['name']:
    arr = df3[df3['name']==dish]['ingredients']
    count = 0
    for ing in arr:
        x = get_nutr(ing)
        if x is not None:
            count = count + x
    df['calorie'] = count
df.head()
```
